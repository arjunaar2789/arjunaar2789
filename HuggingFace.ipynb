```python
from langchain_community.document_loaders import TextLoader
loader=TextLoader('/content/Indian Institute of Technology Madras.txt')
document=loader.load()
print(document)
```

```python
import textwrap
#Preprocessing
def wrap_text_preserve_newlines(text,width=110):
  #Splitting input lines based on newline characters
  lines=text.split('\n')
  #Wrap each line individually
  wrapped_lines=[textwrap.fill(line,width=width) for line in lines]
  #Joining the wrapped lines back together using newline characters
  wrapped_text='\n'.join(wrapped_lines)
  return wrapped_text
```

```python
print(wrap_text_preserve_newlines(str(document[0])))
```

```python
#Text Splitting
from langchain.text_splitter import CharacterTextSplitter
text_splitter=CharacterTextSplitter(chunk_size=1061,chunk_overlap=0)
docs=text_splitter.split_documents(document)
print(len(docs))
print(docs[0])
```

```python
#Embedding
from langchain.embeddings import HuggingFaceEmbeddings
from langchain.vectorstores import FAISS
embeddings=HuggingFaceEmbeddings()
db=FAISS.from_documents(docs,embeddings)
query='Where is IIT Madras?'
doc=db.similarity_search(query)
print(wrap_text_preserve_newlines(str(doc[0].page_content)))
```
